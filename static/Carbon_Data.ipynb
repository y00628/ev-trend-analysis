```python
import pandas as pd
```

```python
tmp = pd.read_csv('table1.csv', skiprows=4)
tmp.head()
```

```python
carbon = tmp.iloc[:,:-4].dropna()
carbon = carbon.iloc[:-1].set_index('State')
carbon
```

```python
# carbon.to_csv('carbon_cleaned.csv')
```

```python
carbon = carbon.transpose()
carbon.astype(float)
```

```python
carbon = carbon.astype(float)
```

```python
carbon.shape[1]
```

```python
carbon['Total Average'] = carbon.sum(axis=1) / 51
```

```python
carbon = carbon.rename('Texas': 'TX', axis=1)
carbon = carbon[['TX', 'Total Average']]
carbon.columns.name = None
carbon = carbon.reset_index().rename({'index': 'Year'}, axis=1)
carbon
```

```python
carbon.to_csv('correlation.csv')
```
