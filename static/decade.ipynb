```python
import numpy as np
import pandas as pd
```

```python
disasters = pd.read_csv('disasters_cleaned_alt.csv')
disasters
```

```python
disasters = disasters[disasters['year'] >= 1960]
```

```python
disasters.insert(0, 'decade', list(map(lambda x: int((x//10)*10), disasters['year'])))
disasters
```

```python
disasters = disasters[~disasters.isna().any(axis=1)].drop('year', axis=1)
```

```python
disasters.set_index('decade', inplace=True)
```

```python
disasters.reset_index(inplace=True)
```

```python
disasters
```

```python
pivot_df = disasters.pivot_table(index=['decade', 'state'], columns='incident_type', aggfunc='size', fill_value=0)

pivot_df['Total'] = pivot_df.sum(axis=1)

reset_df = pivot_df.reset_index()

reset_df
```

```python
reset_df.to_csv('disasters_by_decade.csv', index=False)
```
