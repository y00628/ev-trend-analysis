```python
import pandas as pd
```

```python
disasters = pd.read_csv('disasters_cleaned_alt.csv')
disasters_tx = disasters[(disasters['year'] >= 2002) & (disasters['state'] == 'TX')]
disasters_tx['disaster'] = 1
disasters_tx = disasters_tx.groupby(['year', 'state']).count().reset_index()
disasters_tx = disasters_tx[['year', 'disaster']].rename({'disaster': 'TX', 'year': 'Year'}, axis=1)
disasters_tl = disasters[(disasters['year'] >= 2002)]
disasters_tl['disaster'] = 1
disasters_tl = disasters_tl.groupby(['year', 'state']).count().reset_index()
disasters_tl = disasters_tl.drop('state', axis=1).groupby('year').mean().reset_index()
disasters_tl = disasters_tl[['year', 'incident_type']]
disasters_tl = disasters_tl.rename({'incident_type': 'Total Average', 'year': 'Year'}, axis=1)
```

```python
combined = disasters_tx.merge(disasters_tl, how='left', left_on='Year', right_on='Year')
combined
```
